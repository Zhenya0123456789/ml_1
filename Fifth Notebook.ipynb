```python
file_path = "/content/drive/MyDrive/info.txt"
with open(file_path, "w") as f:
  f.write("Этот файл был создан в Google Colab.\n")
  f.write("Я Женя, по знаку зодиака весы.")
print("Файл успешно сохранен в Google Drive.")

with open(file_path,"r") as f:
  content = f.read()
  print(f"Содержимое файла:\n{content}")

members = [["Novikov E.A",19,"son"],
                 ["Novikov V.A",16,"son"],
                 ["Novikov A.A",12,"son"],
                 ["Novikov A.A",43,"father"],
                 ["Novikova A.V",42,"mother"]]

csv_path = "/content/drive/MyDrive/myfamily.csv"
with open(csv_path, "w") as f:
  for member in members:
    f.write(",".join(map(str,member))+"\n")

print("Файл сохранен")
```
